# DataFrame Slicing
- ##### condition based
- ##### label based
- ##### int index based

```python
import pandas as pd
```

```python
df=pd.read_csv('f:/dataset/analysis/restaurant.csv')
df
```

```python
#condition based
df[df.time=='Dinner']
```

```python
df[(df.time=='Dinner') & (df.gender=='Female')]
```

```python
df
```

```python
#df[(df.day=='Sun')|(df.day=='Sat')]
df[df.day.isin(['Sun','Sat'])]
```

```python
#df[(df.total_bill>=30) & (df.total_bill<=40)]
df[df.total_bill.between(30,40)]
```

```python
#label(named) slicing
#df.loc[startRowIndex:endRowIndex:step,startCol_Label:endCol_label:step]
#Note-->start,end both are inclusive
df.loc[0:4:1,'total_bill':'gender':1]
```

```python
df
```

```python
#int index(position) slicing
#df.iloc[startRowInt_Index:endRowInt_Index:step,startColInt_Index:endColInt_Index:step]
#Note-->start is inclusive but end is exclusive
df.iloc[0:4:1,0:2:1]
```

```python
df.iloc[0:4:1,0:2:1]
```

```python
df.iloc[0:4:1,0:6:1]
```

```python
df.iloc[0:4:1,0:-1:1]
```

```python
df.iloc[:10,1:-1]
```

```python
df.iloc[100:,2:]
```
