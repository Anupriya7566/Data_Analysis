#### df.pivot_table()
- It returns a new df after transforming df according to given parameters
- Generally,we use pivot_table to aggregate(sum,count,mean,min,max,..) a numeric column according to a categorical column.
- we put categorical column on index and column to be aggregated on values,further we define aggregate fun,margins(Total)
- we can also put one more categorical column to column param to further divide data.
- #### df=pd.pivot_table(index='catg_col',column='catg_col',values='num_col',aggfunc='list of operations')

```python
df.pivot_table(index='day',values='total_bill',aggfunc='count',margins=True)
```

```python
df.pivot_table(index='day',values='total_bill',aggfunc=['count','sum','min','max'])
```

```python
df.pivot_table(index='day',columns='time',values='total_bill',aggfunc=['count','sum','min','max'])
```

```python
df.pivot_table(index='day',columns='time',values='total_bill',aggfunc=['count','sum','min','max'],margins=True)
```

```python
df.pivot_table(index='day',columns='time',values='total_bill',aggfunc=['count','sum'],margins=True,margins_name='Total')
```

```python
df.pivot_table(index='gender',values='total_bill') #default aggfunc='mean'
```

```python
df.pivot_table(index='gender',columns='day',values='total_bill',margins=True)
```

```python
df.pivot_table(index=['gender','day','time'],values='total_bill',margins=True)
```

```python
df.pivot_table(index=['gender','day'],columns='time',values='total_bill',margins=True)
```

```python
df.groupby(by='gender')[['total_bill']].mean()
```

```python
df.groupby(by=['gender','day'])[['total_bill']].mean()
```

```python
df.groupby(by=['gender','day'])[['total_bill']].agg(['sum','mean'])
```

### Data Intergration
- process of combining multiple dataframes either row or column wise
- Row wise
  - concat()
- Column wise
  - join()
  - concat()
  - merge()

```python
df1=pd.DataFrame(data={'eid':[1,2,3],'ename':['a','b','c']})
df2=pd.DataFrame(data={'eid':[11,22,33],'ename':['aa','bb','cc']},index=[2,3,4])
df1
```

```python
df2
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
df3=pd.concat([df1,df2],axis=0)
```

```python
df3
```

```python
df3.loc[2]
```

```python
df3.loc[2,'eid']=10
```

```python
df3
```

```python
df1=pd.DataFrame(data={'eid':[1,2,3],'ename':['a','b','c']})
df2=pd.DataFrame(data={'esal':[11000,22000,33000]},index=[1,2,3])
df1
```

```python
df2
```

```python
df1.join(df2,how='inner') #only matching rows based on indexes
```

```python
df1.join(df2,how='left') #all from left and only matching from right
```

```python
df1.join(df2,how='right') #all from right and only matching from left
```

```python
df1.join(df2,how='outer') #all from both
```

```python
df1.join(df2) #default how='left'
```

```python
pd.concat([df1,df2],axis=1) #defult join 'outer'
```

```python
pd.concat([df1,df2],axis=1,join='inner')
```

```python
df1=pd.DataFrame(data={'eid':[1,2,3],'ename':['a','b','c']})
df2=pd.DataFrame(data={'esal':[11000,22000,33000],'eid':[2,3,4]})
df1
```

```python
df2
```

```python
#how to merge dfs based on a common column
pd.merge(df1,df2,on='eid',how='inner')
```
