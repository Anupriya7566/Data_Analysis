```python
import pymysql
import sqlalchemy
```

```python
#con=sqlalchemy.create_engine('url')
con=sqlalchemy.create_engine('mysql+pymysql://root:root@localhost/1030db')
#df=pd.read_sql('select * from emp',con)
#df=pd.read_sql("select * from emp where dept='hr'",con)
#df=pd.read_sql("select empid,ename,empsal from emp where dept='hr'",con)
df=pd.read_sql("select empid,ename,empsal from emp where dept='hr' order by empsal desc",con)
df
```

```python
#df.to_excel('empdata.xlsx',sheet_name='emp') #cwd
df.to_excel('empdata.xlsx',sheet_name='emp',index=False) #cwd
```

```python
df=pd.read_excel('f:/dataset/analysis/movies_sheets.xlsx')
df
```
