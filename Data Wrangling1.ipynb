```python
import pandas as pd
```

```python
df=pd.DataFrame(data={'P':[56,26,70],'C':[66,49,56],'M':[80,90,99]})
df
```

```python
#Data Manipulation
#updating a column
df.P=df.P+5
df
```

```python
#adding a new row
df.loc['a']=[60,50,70]
df
```

```python
#updating single value of dataframe
df.loc[1,'P']=35
df
```

```python
#adding a new column
df['Avg']=round((df.P+df.C+df.M)/3,2)
df
```

```python
df.loc['b']=[20,30,10,20.00]
df.loc['c']=[40,30,70,46.7]
df
```

```python
def findDiv(avg):
    if avg<30:
        return 'Fail'
    elif avg>=30 and avg<=44:
        return '3rd div'
    elif avg>=45 and avg<=59:
        return '2nd div'
    elif avg>=60:
        return '1st div'

df['Div']=df.Avg.apply(findDiv)
df
```

```python
#how to drop row(s)
df.drop('b')
```

```python
df
```

```python
df.drop('b',inplace=True) #inpplace=True,means make changes in existing df
```

```python
df
```

```python
df.drop([1,'c'],inplace=True)
```

```python
df
```

```python
#how to drop column(s)
df.drop('C',axis=1,inplace=True)
```

```python
df
```

```python
df.drop(['P','Avg'],axis=1)
```

```python
df.drop(2,axis=0)
```
